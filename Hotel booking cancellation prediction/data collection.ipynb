```python
cancel_rate = df.groupby('hotel')['is_canceled'].mean()
```

```python
sns.barplot(data=df, x='hotel', y='is_canceled',hue='hotel')
```

```python
numerical_cols = [
    'lead_time', 'stays_in_weekend_nights', 'stays_in_week_nights',
    'adults', 'children', 'babies', 'previous_cancellations',
    'previous_bookings_not_canceled', 'booking_changes',
    'days_in_waiting_list', 'adr', 'required_car_parking_spaces',
    'total_of_special_requests'
]


plt.figure(figsize=(18, 16))
for i, col in enumerate(numerical_cols, 1):
    plt.subplot(4, 4, i) 
    sns.boxplot(y=df[col])
    plt.title(col)
    plt.tight_layout()

plt.suptitle("Boxplots of Numerical Features", fontsize=16, y=1.02)

```

```python
categorical_cols = ['hotel', 'arrival_date_month', 'meal', 'market_segment', 'distribution_channel',
                    'reserved_room_type', 'assigned_room_type', 'deposit_type', 'customer_type',
                    'reservation_status']
```

```python
for col in categorical_cols:
    
    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, x=col, palette="Set2",hue='hotel', order=df[col].value_counts().index)
    plt.title(f"Countplot of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=90)
```

```python
numerical_cols = ['lead_time', 'arrival_date_week_number', 'arrival_date_day_of_month',
                  'stays_in_weekend_nights', 'stays_in_week_nights', 'adults', 'children',
                  'babies', 'previous_cancellations', 'previous_bookings_not_canceled',
                  'booking_changes', 'days_in_waiting_list', 'adr', 'required_car_parking_spaces',
                  'total_of_special_requests']
```

```python
for col in numerical_cols:
    
    plt.figure(figsize=(8, 4))
    sns.histplot(data=df, x=col, palette="Set2",hue='hotel',kde=True)
    plt.title(f"Countplot of {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.xticks(rotation=90)
```
